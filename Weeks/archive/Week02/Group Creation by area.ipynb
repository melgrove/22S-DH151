```python
import pandas as pd

# bring in the roster
df = pd.read_csv('rosterarea.csv')

# replace nans with empty strings
import numpy as np
df = df.replace(np.nan, '', regex=True)

df
```

```python
# create empty list
keywords = []

# loop through rows, and key columns and append to list
for index, row in df.iterrows():
    keywords.append(row.area1.strip())
    keywords.append(row.area2.strip())
    keywords.append(row.area3.strip())
    keywords.append(row.area4.strip())
    keywords.append(row.area5.strip())
    keywords.append(row.area6.strip())

# get rid of nan's
keywords = [x for x in keywords if str(x) != '']

# put it in a dataframe
keywords = pd.DataFrame(keywords,columns=['keyword'])

# add a count 
keywords = keywords.value_counts('keyword')

# reset the table
keywords = keywords.reset_index()

# rename columns
keywords.columns = ['keyword','count']

# show the keyword counts
keywords
```

```python
for index_keywords,keyword in keywords.iterrows():
    print(keyword.keyword.upper())
    print('')
    for index, row in df.iterrows():
        if row.area1.strip() == keyword.keyword or row.area2.strip() == keyword.keyword or row.area3.strip() == keyword.keyword or row.area4.strip() == keyword.keyword or row.area5.strip() == keyword.keyword or row.area6.strip() == keyword.keyword:
            print('    -'+ row['name'].title())
    print('')
        
```
